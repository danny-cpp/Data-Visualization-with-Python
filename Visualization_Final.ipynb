```python
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
# Import data with 1st column as index
df_main = pd.read_csv('Topic_Survey_Assignment.csv', index_col=0)
print('Import sucessful')
```

```python
df_main.head()
```

<h3>Question 2<h3>

```python
# Sort data frame
df_main_sorted = df_main.sort_values(by = ['Very interested'], ascending=False) 

# Convert to percentage, round 2 decimals
df_main_sorted = (df_main_sorted/2233 * 100).round(decimals = 2)
df_main_sorted.head()

bar_graph = df_main_sorted.plot(kind='bar', 
                                figsize = (20,8), 
                                width = 0.8, 
                                color = ["#5cb85c", "#5bc0de", "#d9534f"])

plt.legend(prop={'size': 14})
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title("Percentage of Respondents' Interest in Data Science Areas", fontsize=16)
bar_graph.spines['top'].set_visible(False)
bar_graph.spines['right'].set_visible(False)
bar_graph.spines['left'].set_visible(False)

for p in bar_graph.patches:
    bar_graph.annotate(str(p.get_height()), (p.get_x() * 1, p.get_height() * 1.005))
```
